```python
## updated 27.11.2019
import pandas as pd
import numpy as np

path = 'misc/'
#df = pd.DataFrame.from_csv(path+'chem_py.csv', header=0, sep='\t') 
df = pd.read_csv(path+'chem_py.csv', header=0, sep='\t') 

uni = [u for u in df['Uni']]
rank = [int(e) for e in df['Rang']]

t = zip(uni, rank)

dic = {}
for uni, rank in t:
    dic[uni] = rank

def r(U):
    print(dic.get(U)) 
```

```python
r('Arizona State University')
```

### Wieso Programmiersprache Python?

```python
%matplotlib inline

## Import used packages
import seaborn as sns
sns.set(style="ticks")

## Load data
df = sns.load_dataset("anscombe")

## Visualize results of a linear regression with four datasets
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```

```python
## Import used packages
import networkx as nx
import matplotlib.pyplot as plt

## Create empty directed graph
B=nx.Graph()
## Create list of edges
edges=[(1,2),(1,3),(1,4),(5,6),(5,7),(5,8),(1,9),(9,5)]
## Add edges to graph
B.add_edges_from(edges)
## Draw and show network
nx.draw_networkx(B,node_color='red')
plt.show()
```
